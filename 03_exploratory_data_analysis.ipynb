# Univariate Analysis - Kazakhstan

```python
fig, axes = plt.subplots(1, figsize=(15, 10))

axes[0, 0].hist(kz_data['GPA'], bins=30, edgecolor='black', alpha=0.7, color='steelblue')
axes[0, 0].set_title('Kazakhstan GPA Distribution', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('GPA')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].axvline(kz_data['GPA'].mean(), color='red', linestyle='--', 
                    label=f'Mean: {kz_data["GPA"].mean():.2f}')
axes[0, 0].axvline(kz_data['GPA'].median(), color='green', linestyle='--', 
                    label=f'Median: {kz_data["GPA"].median():.2f}')
axes[0, 0].legend()

```
